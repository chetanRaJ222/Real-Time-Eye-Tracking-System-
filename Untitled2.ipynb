```python
import cv2
import mediapipe as mp
import numpy as np

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(
    refine_landmarks=True,    # IMPORTANT for iris tracking
    max_num_faces=1
)

# Open webcam
cap = cv2.VideoCapture(0)

# Landmark IDs for eyes & iris
LEFT_EYE = [33, 133]
RIGHT_EYE = [362, 263]
LEFT_IRIS = [474, 475, 476, 477]
RIGHT_IRIS = [469, 470, 471, 472]

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    h, w, _ = frame.shape

    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = face_mesh.process(rgb)

    if result.multi_face_landmarks:
        for face_landmarks in result.multi_face_landmarks:

            # Left iris center
            left_iris = np.array([
                [int(face_landmarks.landmark[i].x * w),
                 int(face_landmarks.landmark[i].y * h)]
                for i in LEFT_IRIS
            ])
            left_center = np.mean(left_iris, axis=0).astype(int)

            # Right iris center
            right_iris = np.array([
                [int(face_landmarks.landmark[i].x * w),
                 int(face_landmarks.landmark[i].y * h)]
                for i in RIGHT_IRIS
            ])
            right_center = np.mean(right_iris, axis=0).astype(int)

            # Draw iris
            cv2.circle(frame, tuple(left_center), 3, (0,255,0), -1)
            cv2.circle(frame, tuple(right_center), 3, (0,255,0), -1)

            # Eye corners
            left_eye = [face_landmarks.landmark[i] for i in LEFT_EYE]
            right_eye = [face_landmarks.landmark[i] for i in RIGHT_EYE]

            lx1 = int(left_eye[0].x * w)
            lx2 = int(left_eye[1].x * w)
            rx1 = int(right_eye[0].x * w)
            rx2 = int(right_eye[1].x * w)

            # Gaze ratio
            left_ratio = (left_center[0] - lx1) / (lx2 - lx1)
            right_ratio = (right_center[0] - rx1) / (rx2 - rx1)
            gaze = (left_ratio + right_ratio) / 2

            if gaze < 0.4:
                direction = "Looking Left"
            elif gaze > 0.6:
                direction = "Looking Right"
            else:
                direction = "Looking Center"

            cv2.putText(frame, direction, (30,50),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)

    cv2.imshow("Real-Time Eye Tracking", frame)

    if cv2.waitKey(1) & 0xFF == 27:  # ESC to exit
        break

cap.release()
cv2.destroyAllWindows()
```
